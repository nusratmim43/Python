```python
# QS no 1 :

# here both input convert into float so that the division can be provide smooth solution.
num1 = float(input("Enter 1st number : "))
num2 = float(input("Enter 2nd Number : "))

#Calculation And Print the Output
Addition = num1 + num2
print("Addition :",Addition )
substruction = num1 - num2
print("substruction :",substruction)
Multiplication = num1*num2
print("Multiplication : ",Multiplication)
Division = num1 / num2
print("Division : ",Division);
```
