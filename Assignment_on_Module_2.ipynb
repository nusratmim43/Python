```python
#question 1
name = input("Enter Name : ")
print("Title Case:", name.title())
print("First Character:",name[0])
print("Last Character :",name[-1])
print("Reverse Name:",name[::-1])
print("Number Of 'a' Character: ",name.lower().count('a'))
print("Without Spaces:",name.replace(" ",""))
```

```python
#question No 2
s = "Learning Python Is Fun"
print("Lower Case :",s.lower())
print("Replace Fun : ",s.replace("Fun","Awsome"))
print("3rd to 10th Character:",s[2:10])
print("End with 'Fun'?",s.endswith("Fun"))
print("List of words:",s.split())
```

```python
# Q3: Evaluate expressions

# 5 + 3 * 4
# According to order of operations multiplication first:
# 3*4 = 12, then 5+12 = 17
print("5 + 3 * 4 =", 5 + 3 * 4)

# (5 + 3) * 4
# Parentheses first: 5+3=8, then multiply by 4: 8*4=32
print("(5 + 3) * 4 =", (5 + 3) * 4)

# 100 / 6 and 100 // 6
# 100 / 6 => normal division (float result) => 16.666666...
# 100 // 6 => floor division (integer result) => 16
print("100 / 6 =", 100 / 6)
print("100 // 6 =", 100 // 6)

# 2 ** 5 * 3
# Exponentiation first: 2**5 = 32, then multiply by 3: 32*3=96
print("2 ** 5 * 3 =", 2 ** 5 * 3)
```
